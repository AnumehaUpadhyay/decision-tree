```python
#Label Encoding the class attribute
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(df["deposit"])
df["deposit"] = label_encoder.transform(df["deposit"])
```

```python
df['bal_sign'] = pd.Series(df['balance'] > 0)
```

```python
labels = ['housing', 'default', 'loan']
for label in labels:
    label_encoder = LabelEncoder()
    label_encoder.fit(df[label])
    df[label] = label_encoder.transform(df[label])
```

```python
cols = ['job', 'contact', 'marital','education', 'poutcome', 'month', 'day']
for label in cols:
    label_encoder = LabelEncoder()
    label_encoder.fit(df[label])
    df[label] = label_encoder.transform(df[label])
```

```python
df = pd.get_dummies(df, columns = cols)
```

```python
df.head(2)
```

```python
labels = df.deposit
scaler = MinMaxScaler()
features = pd.DataFrame(scaler.fit_transform(df.drop(['deposit'], axis = 1)), columns = df.drop(['deposit'], axis = 1).columns)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=1, stratify = labels)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)
# fit the model
clf_gini.fit(X_train, y_train)
```

```python
y_pred_gini = clf_gini.predict(X_test)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(clf_gini,filled=True)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred_gini))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred_gini))
```

```python
cm=confusion_matrix(y_test,y_pred_gini)
print(cm)
```
